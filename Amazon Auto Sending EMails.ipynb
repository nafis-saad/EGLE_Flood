```python

import win32com.client
import pandas as pd

email_column = "Respondent_Email"  # Replace with the exact column name
name_column = "Respondent_Name"  # Replace with the exact column name

# Ensure column names are consistent
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces

# Path to the logo image
logo_path = "D:/Flood_Project/Amazon_Gift_Card/wsu.png"  # Update this with the actual path of the image on your system

# Initialize Outlook
outlook = win32com.client.Dispatch("Outlook.Application")
```

```python
# Loop through the dataframe to send emails
for index, row in df.iterrows():
    recipient_email = row[email_column]
    recipient_name = row[name_column]

    mail = outlook.CreateItem(0)  # Create a new email
    mail.To = recipient_email
    mail.CC = "xdqian@wayne.edu"  # Add CC

    # Customize the email subject
    mail.Subject = "🎉 Thank You for Participating in Our Study! 🎉"

    # Create the HTML email body with the logo
    mail.HTMLBody = f"""
    <html>
    <body style="font-family: Arial, sans-serif; color: #333;">
    
    <p>Dear <strong>{recipient_name}</strong>,</p>

    <p>We hope this email finds you well. You may recall participating in our <strong>flooding impact study</strong> last August, 
    sponsored by the <strong>Michigan Department of Environment, Great Lakes, and Energy (EGLE)</strong> and the 
    <strong>National Oceanic and Atmospheric Administration (NOAA)</strong>. We sincerely appreciate the time and effort 
    you dedicated to completing our survey on the impact of flooding on travel in Detroit.</p>

    <p><strong>Great News! 🎉</strong></p>
    <p>After reviewing all survey responses, we are pleased to inform you that you have been <strong>randomly selected</strong> 
    to receive a <strong>$10 Amazon gift card</strong> as a token of our appreciation. The selection was made from a pool of 
    qualified participants who met the study criteria.</p>

    <p><strong>How to Claim Your Gift Card:</strong></p>
    <ul>
        <li>Please check your email: <strong>{recipient_email}</strong> (including spam/junk folder).</li>
        <li>If you have not received it, please contact us at <a href="mailto:nafis_fuad@wayne.edu">nafis_fuad@wayne.edu</a>.</li>
    </ul>

    <p>Once again, <strong>thank you</strong> for your valuable contribution to this important research. 
    Your participation plays a crucial role in helping us and our partners develop better mobility solutions for 
    Detroit residents during flooding events.</p>

    <p>Best regards,</p>

    <p><strong>Nafis</strong>,<br>
    <em>On behalf of Dr. Xiaodong Qian</em><br>
    Assistant Professor<br>
    Department of Civil & Environmental Engineering<br>
    Wayne State University (WSU)<br>
    Detroit, MI, 48202<br>
    Office Tel: 313-577-3854<br>
    Email: <a href="mailto:xdqian@wayne.edu">xdqian@wayne.edu</a></p>

    <img src="cid:wayne_state_logo" alt="Wayne State University Logo" style="width:200px; margin-top:10px;">

    </body>
    </html>
    """

    # Add the logo as an attachment
    attachment = mail.Attachments.Add(logo_path)
    attachment.PropertyAccessor.SetProperty("http://schemas.microsoft.com/mapi/proptag/0x3712001F", "wayne_state_logo")

    # Send the email
    mail.Send()
    
print("Emails sent successfully!")
```
